**INFORMASI FITUR**

**1. AREA atau DAERAH**

Fitur ini mengukur luas dari objek. Luas dapat dihitung dalam unit piksel atau unit luas lainnya, tergantung pada resolusi data. Luas memberikan informasi tentang ukuran relatif objek. Data yang terkait dengan fitur ini memiliki tipe data numerik.

**2. PERIMETER**

Perimeter mengukur panjang garis batas objek. Ini diukur sebagai jumlah panjang semua tepi objek. Perimeter bisa memberikan indikasi seberapa kompleks bentuk objek tersebut. Data yang terkait dengan fitur ini memiliki tipe data numerik.

**3. MAJOR AXIS LENGHT atau PANJANG SUMBU UTAMA**

Sumbu utama adalah sumbu terpanjang dalam elips yang mengelilingi objek. Panjang sumbu ini memberikan gambaran tentang dimensi utama objek dan arah orientasi elips. Data yang terkait dengan fitur ini memiliki tipe data numerik.

  **4. MINOR AXIS LENGHT atau PANJANG SUMBU KECIL**

  Sumbu kecil adalah sumbu terpendek dalam elips yang mengelilingi objek. Ini memberikan informasi tentang dimensi kedua objek dan dapat membantu menggambarkan bentuk elips. Data yang terkait dengan fitur ini memiliki tipe data numerik.

- **Panjang Sumbu Utama (Biru):**

  Panjang ini adalah jarak terpanjang dari ujung ke ujung pada biji beras. Dalam konteks deteksi tepi, L menggambarkan jarak terpanjang antara dua titik sepanjang kontur yang terdeteksi.

- **Panjang Sumbu Kecil (Merah):**

  Panjang ini adalah jarak terpendek yang tegak lurus terhadap sumbu utama. Dalam konteks deteksi tepi, W menggambarkan jarak terpendek antara dua titik yang tegak lurus terhadap sumbu utama.

**5. ECCENTRICITY atau EKSENTRISITAS**

Eksentrisitas mengukur sejauh mana elips yang mengelilingi objek mendekati bentuk lingkaran. Nilai eksentrisitas 0 menunjukkan objek yang bentuknya mendekati lingkaran sempurna, sementara nilai mendekati 1 menunjukkan elips yang sangat panjang. Data yang terkait dengan fitur ini memiliki tipe data numerik.

**6. CONVEX AREA atau DAERAH CEMBUNG**

Luas cembung mengukur luas daerah yang diukur dari cembung (convex hull) objek. Cembung adalah poligon terkecil yang dapat mencakup seluruh objek. Luas ini memberikan informasi tentang sejauh mana objek dapat dianggap "cembung". Data yang terkait dengan fitur ini memiliki tipe data numerik.

**7. EXTENT atau LUAS**

Extent adalah rasio antara luas objek dan luas kotak terkecil yang dapat mengelilingi objek. Nilai 1 menunjukkan objek yang mengisi kotak dengan sempurna, sementara nilai yang lebih rendah menunjukkan objek yang mungkin memiliki bentuk yang lebih tidak teratur. Data yang terkait dengan fitur ini memiliki tipe data numerik.

**8. CLASS atau KELAS**

Kelas adalah label kategori atau jenis keanggotaan dari objek. Ini adalah informasi klasifikasi yang menunjukkan keberadaan objek dalam kategori tertentu, seperti 0 : kelas "Rice Cammeo" atau 1 : kelas "Rice Osmancik". **LINEAR DISCRIMINANT ANALYSIS**

Linear Discriminant Analysis adalah metode statistik dan machine learning yang digunakan untuk pemisahan kelas dalam masalah klasifikasi. Tujuannya adalah untuk mencari transformasi linear dari data asli ke ruang dimensi yang lebih rendah sehingga kelas data terpisah dengan baik. Memiliki dua kelas data, yang disebut kelas 1 dan kelas 0. LDA berusaha untuk memproyeksikan data ini ke satu dimensi (barisan) agar perbedaan antara kelas menjadi masksimal.

Langkah-langkah utama dalam LDA adalah sebagai berikut :

- Menghitung vektor rata-rata untuk masing-masing kelas. Misalkan µ_1 dan µ_0 adalah vektor rata-rata untuk kelas 1 dan kelas 0, masing-masing.
- Menghitung matriks perincian dalam (within-class scatter matrix) dan matriks pencirian antar-kelas (between-class scatter matrix).

  1.   Matriks pencirian dalam (S_W) adalah matriks yang mengukur variasi dalam setiap kelas. Ini didefinisikan sebagai jumlah matriks pencirian dalam untuk masing-masing kelas :

    **S_W = Σ (x - µ_i)(x - µ_i)^T**

    di mana x adalah vektor data, µ_i adalah vektor rata-rata kelas i, dan Σ adalah operator penjumlahan untuk semua data dalam kelas.

  2.   Matriks pencirian antar-kelas (S_B) adalah matriks yang mengukur perbedaan antara kelas. Ini didefinisikan sebagai :

    **S_B = (µ_1 - µ_0)(µ_1 - µ_0)^T**

- Menghitung matriks eigen (matriks yang memiliki vektor eigen dan nilai eigen).

- Mengurutkan vektor eigen dalam urutan menurun berdasarkan nilai eigen.

- Memilih sejumlah vektor eigen teratas sesuai dengan jumlah dimensi yang diinginkan untuk proyeksi.

- Membentuk matriks proyeksi (projection matrix) dengan menggabungkan vektor eigen terpilih sebagai kolom-kolom matriks.

- Data dapat diproyeksikan ke ruang barisan (barisan tunggal) dengan mengalikan matriks data asli dengan matriks proyeksi.

Dalam praktiknya, proyeksi data ke ruang barisan tunggal akan memungkinkan Anda untuk mengklasifikasikan data berdasarkan ambang tertentu. Misalnya, jika nilai proyeksi lebih besar dari ambang tertentu, data dapat diklasifikasikan ke dalam kelas 1; jika lebih kecil, data dapat diklasifikasikan ke dalam kelas 0.
